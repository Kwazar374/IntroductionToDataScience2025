## Exercise 2
Using scatter plot:  
a) Show the relation of square footage and price.  
b) Check how price is influenced by apartment's grade. Find other variables correlated with price.  
c) Check how good apartments are distributed over the city. First, plot zipcode versus price. Then, use apartments coordinates (longitude and lattitude) in order to show where apartments of good grade are located.

```python
#a)
#Your code:

plt.scatter(data.sqft_living, data.price)
plt.xlabel("Square footage")
plt.ylabel("Price")

#Hint plt.scatter
```

***The plot above shows that square footage and price are weakly correlated, which suggests that other factors also influence house prices. However, it can still be observed that larger houses tend to cost more.***

```python
#b)
#Your code:

plt.scatter(data.grade, data.price)
plt.xlabel("Grade")
plt.ylabel("Price")
```

***The plot above shows that a higher grade generally corresponds to a higher price. However, similar to the relationship between square footage and price, there are also some inexpensive houses with relatively high grades. Another thing worth noting is that the grade variable takes discrete values, which is why there are distinct vertical lines on the plot.***

```python
corr_matrix = house_numerical_df.corr(numeric_only=True)
corr_with_price = corr_matrix["price"].sort_values(ascending=False)
print(corr_with_price)
```

***The correlation analysis shows that sqft_living (living space) has the highest correlation with price (r = 0.70), followed by sqft_above (space above the ground) and bathrooms. While these variables are the most influential numerically, the correlation is not perfect. This aligns with our earlier observation that there are some inexpensive houses with large square footage, indicating that other factors besides size also affect the price. Most other variables, such as yr_built, sqft_lot, or long, show very weak correlation with price.***

```python
#c)
#Your code:

plt.scatter(data.zipcode, data.price)
plt.xlabel("Zipcode")
plt.ylabel("Price")
```

***The chart above provides an overview of the district boundaries and illustrates the typical house prices in each area.***

```python
#Your code:

plt.figure(figsize=(12, 8), dpi=800)
plt.scatter(data.long, data.lat, c = data.price, cmap = "Greens_r")
plt.xlabel("Longtitude")
plt.ylabel("Latitude")
```

***The scatter plot shows house prices in different areas. Dark points, which represent cheaper houses, are mostly around the edges, while lighter points, which represent more expensive houses, are in the center. This shows that houses in the county center are usually more expensive.***

## Exercise 4
"I am interested in the house with 3 bedrooms, 2 bathrooms and 2 floors"  
a) Compute the basic statistics of apartments fulfilling above constraints.  
b) Distinct accepted apartments on any of previously created plots.  
c) Find such apartments with best price to footage ratio.  

```python
#a)
#Your code:

data_bbf = data.loc [(data['bedrooms'] == 3) & (data['bathrooms'] == 2) & (data['floors'] == 2)]
data_bbf.describe()

#Hint: pandas.DataFrame.loc
```

```python
#b)
#Your code:

plt.scatter(data.grade, data.price)
plt.scatter(data_bbf.grade, data_bbf.price)
plt.xlabel("Grade")
plt.ylabel("Price")
```

```python
#c)
#Your code:

data_bbf = data_bbf.copy()
data_bbf['price_to_footage'] = data_bbf['price'] / data_bbf['sqft_living']

data_bbf.sort_values('price_to_footage', ascending=True)

#Hint: .loc, .sort_values(), .index
```

***The best choice would be apartment with ID 7452500565.***
